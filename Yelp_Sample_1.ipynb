### Create a sample

```python
yelp_reviews=pd.read_csv("review_data.csv")
```

```python
yelp_reviews.head(5)
```

```python
yelp_reviews['year'] = pd.DatetimeIndex(yelp_reviews['date']).year
yelp_reviews['year'].value_counts()
```

```python
yelp_reviews['useful'].value_counts()
```

```python
yelp_reviews['length'] = yelp_reviews['text'].map(len)
```

```python
yelp_reviews['length'].value_counts()
```

```python
yelp_reviews = yelp_reviews[(yelp_reviews.useful == 0) | (yelp_reviews.useful >= 10)]
yelp_reviews["labels"] = yelp_reviews["useful"].apply(lambda x: 1 if x >= 10 else 0)
# reviews=reviews[reviews.stars!=3]
yelp_reviews.head()
```

```python
yelp_reviews['labels'].value_counts()
```

```python
business_cnt = pd.DataFrame(yelp_reviews.groupby('business_id')['business_id'].count())
```

```python
business_cnt.columns = ['business_cnt']
```

```python
business_cnt.reset_index()
```

```python
yelp_reviews = pd.merge(yelp_reviews, business_cnt, on = 'business_id', how = 'inner')
```

```python
yelp_reviews.head(5)
```

```python
# yelp_reviews = yelp_reviews[yelp_reviews['year'] >= 2017]
```

```python
yelp_reviews['labels'].value_counts()
```

```python
# yelp_sample = yelp_reviews[yelp_reviews['year'] == 2019]
yelp_sample1 = yelp_reviews[yelp_reviews['labels'] == 0].sample(10000)
yelp_sample2 = yelp_reviews[yelp_reviews['labels'] == 1].sample(10000)
yelp_sample = pd.concat([yelp_sample1, yelp_sample2], ignore_index=True)
```

```python
yelp_sample['labels'].value_counts()
```

```python
cuisine_type = pd.get_dummies(yelp_sample["cuisine"])
```

```python
cuisine_type.head(5)
```

```python
yelp_sample['cuisine'].value_counts()
```

```python
yelp_sample = yelp_sample.join(cuisine_type)
```

```python
yelp_sample.shape
```

```python
yelp_sample.columns
```

```python
yelp_sample.to_csv('./review_sample_1.csv',index=False, header=True)
```
