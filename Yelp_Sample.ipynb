### Create a sample

```python
yelp_reviews=pd.read_csv("review_data.csv")
```

```python
yelp_reviews.head(5)
```

```python
yelp_reviews['year'] = pd.DatetimeIndex(yelp_reviews['date']).year
```

```python
yelp_reviews['year'].value_counts()
```

```python
yelp_reviews['length'] = yelp_reviews['text'].map(len)
```

```python
yelp_reviews['length'].value_counts()
```

```python
yelp_reviews[yelp_reviews['year'] == 2019]['cuisine'].value_counts()
```

```python
business_cnt = pd.DataFrame(yelp_reviews.groupby('business_id')['business_id'].count())
```

```python
business_cnt.columns = ['business_cnt']
```

```python
business_cnt.reset_index()
```

```python
yelp_reviews = pd.merge(yelp_reviews, business_cnt, on = 'business_id', how = 'inner')
```

```python
yelp_reviews.head(5)
```

```python
yelp_reviews[yelp_reviews['year'] == 2019]['cuisine'].value_counts()
```

```python
yelp_reviews["labels"]= yelp_reviews["useful"].apply(lambda x: 1 if x >= 1  else 0)
```

```python
yelp_sample = yelp_reviews[yelp_reviews['year'] == 2019].sample(n = 20000, random_state=1)
```

```python
yelp_sample[yelp_sample['year'] == 2019]['cuisine'].value_counts()
```

```python
cuisine_type = pd.get_dummies(yelp_sample["cuisine"])
```

```python
cuisine_type.head(5)
```

```python
yelp_sample = yelp_sample.join(cuisine_type)
```
